```python
import pandas as pd
import numpy as np
```

```python
# Task 1 : Create a pandas dataframe (DataFrame name as 'df') with numpy random values (4 features and 4 observation).
```

```python
random_values = np.random.rand(4,4)
df = pd.DataFrame(random_values)
print("\nDataFrame with Random Values :\n")
df
```

```python
# Task - 2 : Rename the task - 1 'df' dataframe column names to 'Random value 1', 'Random value 2', 'Random value 3' & 'Random value 4'.
```

```python
df.columns = ['Random Value 1','Random Value 2','Random Value 3','Random Value 4']
print("\nDataFrame with Renamed Column Names :\n")
df
```

```python
# Task 3 : Find the descriptive statistics of the 'df' dataframe.
```

```python
descriptive_stats = df.describe()
print("\nDescriptive Statistics of the DataFrame :\n")
descriptive_stats
```

```python
# Task 4 : Check for the null values in 'df' and find the data type of the columns.
```

```python
null_values = df.isnull()
print("\nNull Values of Columns :\n")
print("Null Values :\n")
print(null_values)
null_values = df.isnull().sum()
print("\nNull Value Sum of Columns :\n")
print("Null Values :\n")
print(null_values)
data_types = df.dtypes
print("\nData Types :\n")
print(data_types)
```

```python
# Task 5 : Display the 'Random value 2' & 'Random value 3' columns with location method and index location method.
```

```python
random_values_2_3_loc = df.loc[:,'Random Value 2':'Random Value 3']
print("\nDisplaying 'Random Value 2' and 'Random Value 3' columns using Location Method :\n")
random_values_2_3_loc
```

```python
random_values_2_3_iloc = df.iloc[:,1:3]
print("\nDisplaying 'Random Value 2' and 'Random Value 3' columns using Index Location Method :\n")
random_values_2_3_iloc
```
